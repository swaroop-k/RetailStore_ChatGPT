```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading the dataset
retail_data = pd.read_csv(r'.\data\realistic_retail_data.csv')
retail_data.head()
```

```python
# 1. What is the distribution of customer ages?

# describe method
age_distribution = retail_data['Age'].describe()
print(age_distribution)

# Set plot style
sns.set(style="whitegrid")

# 1. Distribution of customer ages
plt.figure(figsize=(10,6))
sns.histplot(retail_data['Age'], bins=30, kde=True)
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. How many customers are there in each gender category?
gender_counts = retail_data['Gender'].value_counts()
print(gender_counts)


# visualization
# Creating a bar plot
# Count the number of customers in each gender category
gender_count = retail_data['Gender'].value_counts().reset_index()
gender_count.columns = ['Gender', 'Count']

# Plotting the barplot
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Gender', y='Count', data=gender_count)

# Adding the values on top of the bars
for index, row in gender_count.iterrows():
    barplot.text(x=row.name, y=row.Count, s=str(row.Count), color='black', ha="center")

# Adding title and labels
plt.title('Number of Customers in Each Gender Category')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')

# Display the plot
plt.show()
```

```python
# 3. What is the average amount spent per transaction?
avg_transaction = retail_data['AmountSpent'].mean()
print(f'Average amount spent per transaction: ${avg_transaction:.2f}')
```

```python
# 4. What are the most common cities among customers?
common_cities = retail_data['City'].value_counts()
common_cities
```

```python
# 5. How many transactions fall into each product category?
product_category_counts = retail_data['ProductCategory'].value_counts()
print(product_category_counts)
```

```python
# 6. What is the distribution of product ratings?
product_rating_distribution = retail_data['ProductRating'].describe()
print(product_rating_distribution)

# visualization
plt.figure(figsize=(10, 6))
sns.histplot(retail_data['ProductRating'], bins=30, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Product Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# 7. How many transactions used each payment method?
payment_method_counts = retail_data['PaymentMethod'].value_counts()
payment_method_counts
```

```python
# 8. What is the percentage of transactions that were returned?
return_percentage = retail_data['IsReturned'].mean() * 100
print(f'{return_percentage:.2f}% of transactions were returned')
```

```python
# 9. What is the distribution of delivery statuses?
delivery_status_counts = retail_data['DeliveryStatus'].value_counts()
delivery_status_counts
```

```python
# 10.What is the average delivery time in days?
average_delivery_time_days = retail_data['DeliveryTimeDays'].mean()
average_delivery_time_days
```

```python
# 11. What is the distribution of delivery fees?
delivery_fee_distribution = retail_data['DeliveryFee'].describe()
print(delivery_fee_distribution)


# visualization
plt.figure(figsize=(10, 6))
sns.histplot(retail_data['DeliveryFee'], bins=30, kde=True)
plt.title('Distribution of Delivery Fees')
plt.xlabel('Delivery Fee')
plt.ylabel('Frequency')
plt.show()
```

```python
# 12. How many feedback scores are there in each category (Positive, Neutral, Negative)?
feedback_score_counts = retail_data['FeedbackScore'].value_counts()
print(feedback_score_counts)
```

```python
# 13. What is the average review length?
average_review_length = retail_data['ReviewLength'].mean()
print("Average review length: ", average_review_length)
```

```python
# 14. How many customers fall into each customer satisfaction category?
customer_satisfaction_counts = retail_data['CustomerSatisfaction'].value_counts()
print(customer_satisfaction_counts)
```

```python
# 15. What is the distribution of transaction dates?
transaction_date_distribution = retail_data['TransactionDate'].describe()
print(transaction_date_distribution)

# visualization
plt.figure(figsize=(10, 6))
sns.histplot(retail_data['TransactionDate'], bins=30, kde=True)
plt.title('Distribution of Transaction Dates')
plt.xlabel('Transaction Date')
plt.ylabel('Frequency')
plt.show()
```
